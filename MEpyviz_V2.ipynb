```python
#Input Layout

#InputLayoutFrame.propagate(False)
#[[1], [0,  2, 3, 4]]

entries=[]
locrows=[]
loccol=[]

group_iN=np.empty([6, 4], dtype=object) #the layour input
count = 80
clk=0

def collect():
    
    
    
    
    global entries
    global locrows
    global loccol
    global group_iN
    
   
    
    for idx in range(len(entries)):
        
        
        row=locrows[idx].get()
        
        row=[int(item) for item in row if row!='Row']
        col=loccol[idx].get()
        col=[int(item) for item in col if col!='Column']
        
        if entries[idx].get()!='Entry':
        
            group_iN[np.ix_( col, row )]=entries[idx].get()
            
    group_iN=group_iN.ravel(order='F')
    print(group_iN)
        

# To keep track of inserted entries
def add():
    global CollectButton
    global count
    global entries
    global locrows
    global loccol
    global clk
    MAX_NUM = 1000# Maximum number of entries
    if clk>0:
        
        CollectButton.destroy()
        
    if count <= MAX_NUM:
        entries.append(Entry(root))
        locrows.append(Entry(root))
        loccol.append(Entry(root))
        
        # Create and append to list
       
        locrows[-1].place(x=700, y=count)#place(relx=0.2, rely=count)
        locrows[-1].insert(0, 'Row')
        loccol[-1].place(x=800, y=count)#place(relx=0.4, rely=count)
        loccol[-1].insert(0, 'Column')
        entries[-1].place(x=950, y=count)#.place(relx=0.6, rely=count)
        entries[-1].insert(0, 'Entry')
        
        # Place the just created widget
        count += 30
        CollectButton=Button(root, text='Collect', command=collect)
        CollectButton.place(x=700, y=count+20)
        clk=clk+1# Increase the count by 1 
   
        
                
                       
Button(root, text='Layout Entry', command=add).place(x=700, y=50)




```
